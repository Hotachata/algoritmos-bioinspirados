Práctica de Inteligencia de Enjambre

Realizar la minimización de la función x^2 + y^2 + [25 * senx + seny]
en el intervalo de valores (-5,5) para (x, y) usando PSO.

Versión de PSO: Global
Número de partículas: 20
Iteraciones: 50
a= 0.8    (inercia)
b1= 0.7  (aprendizaje local)
b2 = 1.2   (aprendizaje global)

En cada iteración imprimir:
Posición de la partícula
Velocidad
pbest
gbest

```python
import random
import math

# Cada particula tiene una posición x, y
def random_distancia():
    return [round(random.uniform(-5, 5), 2), round(random.uniform(-5, 5), 2)]

# Crear una particula inicial
def una_particula_inicial():
    posicion = random_distancia()  # Crea 'x' y 'y'
    velocidad_actual = [0, 0]  # Define la velocidad actual como una lista con dos valores
    pbest = posicion  # La mejor posición es la inicial
    pbest_v = [0, 0]
    nueva_velocidad = random_distancia()  # Genera una velocidad aleatoria
    fitness = aptitud(posicion)  # Función de evaluación
    return [posicion, velocidad_actual, pbest, nueva_velocidad, fitness, pbest_v]

# Función de aptitud
def aptitud(d):
    x = d[0]
    y = d[1]
    return round(x**2 + y**2 + (25 * (math.sin(x) + math.sin(y))), 2)

# Actualización de la velocidad
def una_nueva_velocidad(velocidad_actual, posicion, pbest, gbest):
    nueva_velocidad = []
    r1 = random.uniform(0, 1)
    r2 = random.uniform(0, 1)
    for i in range(2):
        nueva_v = round((0.8 * velocidad_actual[i]) + (0.7 * r1 * (pbest[i] - posicion[i])) + (1.2 * r2 * (gbest[i] - posicion[i])), 2)
        nueva_velocidad.append(nueva_v)
    return nueva_velocidad

# Actualización de la posición de una partícula
def una_particula_iteracion(posicion, velocidad_actual, pbest, nueva_velocidad, gbest, pbest_v):
    posicion = [round(posicion[0] + nueva_velocidad[0], 2), round(posicion[1] + nueva_velocidad[1], 2)]
    velocidad_actual = nueva_velocidad
    if aptitud(posicion) < aptitud(pbest):
        pbest = posicion
        pbest_v = velocidad_actual
    nueva_velocidad = una_nueva_velocidad(velocidad_actual, posicion, pbest, gbest)
    fitness = aptitud(posicion)
    return [posicion, velocidad_actual, pbest, nueva_velocidad, fitness, pbest_v]

# Función para crear una posición enjambre
def ejambrar(pbest, velocidad_actual):
    x = pbest[0] + velocidad_actual[0]
    y = pbest[1] + velocidad_actual[1]
    posicion = [round(x, 2), round(y, 2)]
    aptitud_p = aptitud(posicion)
    return [posicion, aptitud_p]

# Configuración de parámetros
n_particulas = 20
n_iteraciones = 50
iteracion = []
gbest = None
gbest_score = float('inf')
pbest_vector = [[0, 0] for _ in range(n_particulas)]
pbest_vector_v = [[0, 0] for _ in range(n_particulas)]

# Ciclo principal de iteraciones
for j in range(n_iteraciones):
    if j == 0:
        for i in range(n_particulas):
            particula = una_particula_inicial()
            iteracion.append(particula)
            if particula[4] < gbest_score:
                gbest = particula[2]
                gbest_score = particula[4]
        print(f"GBest inicial: {gbest}")
    else:
        auxiliar_vector = []
        print(f"\nIteración {j}:")
        for i in range(n_particulas):
            particula = iteracion[i]
            nueva_particula = una_particula_iteracion(particula[0], particula[1], particula[2], particula[3], gbest, particula[5])
            auxiliar_vector.append(nueva_particula)
            if nueva_particula[4] < gbest_score:
                gbest = nueva_particula[2]
                gbest_score = nueva_particula[4]

            # Guardando el mejor `pbest` para el enjambre
            if j == n_iteraciones - 1:
                pbest_vector[i] = nueva_particula[2]
                pbest_vector_v[i] = nueva_particula[1]

            # Imprimir valores requeridos para cada partícula
            print(f"  Partícula {i + 1}:")
            print(f"    Posición: {nueva_particula[0]}")
            print(f"    Velocidad: {nueva_particula[1]}")
            print(f"    pbest: {nueva_particula[2]}")
            print(f"    gbest: {gbest}")

        # Crear el enjambre al final de las iteraciones
        if j == n_iteraciones - 1:
            enjambre = []
            print("\nEnjambre generado:")
            for i in range(n_particulas):
                miembro = ejambrar(pbest_vector[i], pbest_vector_v[i])
                enjambre.append(miembro)
                print(f"  Miembro {i + 1}: {miembro}")

        iteracion = auxiliar_vector  # Actualizar la lista de partículas con la nueva iteración
```
